## `Dispersion` : How much the values are far apart from the `mean or average or central tendency` of the whole data.

### `Variance (σ)` is one way of calculating the Dispersion : (x-x̄)^2 :`x` is the distance from mean : `x̄` is the mean.

`We donot take absolute in variance(x-x̄) : because of the derivative resaons (absolute has no dariative) or Calculus. So`

`We take square to remove the negative from the negative distance values`.
`Thus now varience has a unit "meter-square"`. 

To solve this we use

### `"Standard Daviation" : which is the square-root of variance(σ).` : Square root of the average square distances of values from mean.

#### How far are the values from the mean.

# Generating a lot of numbers

```python
numSamples = 10000
```

```python
uniform_values = np.random.uniform(low=0.0, high=10.0 ,size = numSamples) 
```

```python
sns.histplot(uniform_values , bins = 20 , kde = False)
plt.ylabel("frequency")
plt.xlabel("values")
plt.title("Uniform values")
sns.despine(offset = 10 ,trim = True) #move axis away
plt.show()
```

```python
normal_values = np.random.normal(loc = 5.0, scale = 10.0 ,size = numSamples) #more values close to "loc" 
```

```python
sns.histplot(normal_values , bins = 20 , kde = False)
plt.ylabel("frequency")
plt.xlabel("values")
plt.title("normal values")
sns.despine(offset = 10 ,trim = True) #move axis away
plt.show()
```

### Look how they( `the uniform ones` & `Normal values mostly around 5`) both have almost same mean() : central tendency.

```python
print("Uniform values mean : ", np.mean(uniform_values))
```

```python
print("Normal values mean : ", np.mean(normal_values))
```

### Therefore Variance is required!!!

```python
print("Uniform values varience :", np.var(uniform_values))
```

```python
print("Normal values varience :",np.var(normal_values))
```

### As they are squared values So , We have Standard Daviation

```python
print("Uniform values standard daviation :", np.std(uniform_values))
print("Uniform values standard daviation :", np.std(normal_values))
```

```python
normal_values = np.random.normal(loc=5.0, scale=10.0, size = numSamples )
nv_mean = np.mean(normal_values)
nv_sd = np.std(normal_values)
```

```python
sns.histplot(normal_values,bins = 20 , kde=False)
plt.xlim(-60,60)                        #!!!
plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("Normal values")
# sns.despine(offset=10,trim=True)       #move axis away

#plot the SD line
x_c , y_c = ([nv_mean , nv_mean] , [0,1500])
plt.plot(x_c,y_c,color='green',linewidth=2)

x_c , y_c = ([nv_mean + nv_sd , nv_mean + nv_sd] , [0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

x_c , y_c = ([nv_mean-nv_sd , nv_mean-nv_sd] , [0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

plt.show()
```

```python
normal_values_2 = np.random.normal(loc=5.0, scale=15.0, size = numSamples )
nv_mean_2 = np.mean(normal_values_2)
nv_sd_2 = np.std(normal_values_2) 
```

```python
sns.histplot(normal_values_2,bins = 20 , kde=False)
plt.xlim(-60,60)                        #!!!
plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("Normal values_2")
# sns.despine(offset=10,trim=True)       #move axis away

#plot the SD line
x_c , y_c = ([nv_mean_2 , nv_mean_2] , [0,1500])
plt.plot(x_c,y_c,color='green',linewidth=2)

x_c , y_c = ([nv_mean_2 + nv_sd_2 , nv_mean_2 + nv_sd_2] , [0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

x_c , y_c = ([nv_mean_2 - nv_sd_2 , nv_mean_2 -nv_sd_2] , [0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

plt.show()
```

#### Lets put them togather

```python
normal_values = np.random.normal(loc=5.0, scale=10.0, size = numSamples )
normal_values_2 = np.random.normal(loc=5.0, scale=15.0, size = numSamples)
#make a function : the values passed will deside the plot : n_v or nv_2
def plot_hist(values,label):
    nv_mean = np.mean(values)
    nv_sd = np.std(values)

    sns.histplot(values,bins=20,kde=False)

    plt.xlim(-60,60)

    plt.ylabel("Frequency")
    plt.xlabel("Values")
    plt.title(label)

    #plot the SD line
    x_c , y_c = ([nv_mean_2 , nv_mean_2] , [0,1500])
    plt.plot(x_c,y_c,color='green',linewidth=2)

    x_c , y_c = ([nv_mean_2 + nv_sd_2 , nv_mean_2 + nv_sd_2] , [0,1500])
    plt.plot(x_c,y_c,color='red',linewidth=2)

    x_c , y_c = ([nv_mean_2 - nv_sd_2 , nv_mean_2 -nv_sd_2] , [0,1500])
    plt.plot(x_c,y_c,color='red',linewidth=2)

    plt.show()

plot_hist(normal_values,"Less Varient")
plot_hist(normal_values_2,"More Varient")
```
