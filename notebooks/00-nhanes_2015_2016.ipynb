# Data Visualiation

```python
y = [6,7,2,4,5] #Example plot
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y) #plotting data like this is a bad idea.We must add x,y labels and title.
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(y)

plt.title("First Example Plot")
plt.xlabel("time(s)")
plt.ylabel("distance(m)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,5)) #plotting the actual nhanes data.

x = range(da['BMXWT'].size)

y = da['BMXWT']

plt.scatter(x,y, marker='x', s=0.5) #scatter removes the line between points. marker is to set the size of dots.
plt.title("Weight of Individuals")
plt.xlabel("ID of Individuals") 
plt.ylabel("Weight(Kg)")

plt.show()
```
