# One Hot Encoding

```python
B = ['bird','cat','dog','bird','bird']

d = {'categorical' : B}

df = pd.DataFrame(d)
```

### dummies is used to create columns corresponding to unique values


```python
dfDummies = pd.get_dummies(df['categorical'], dtype = int, prefix = 'category')
```
